### Pergunta
3. A ordem da remoção das stopwords e do stemmming pode gerar problema?
4. Ao aplicar stemming com RSLPStemmer, que tipo de transformação ocorre nas palavras? As formas resultantes são linguisticamente válidas?

```
# Remover stopwords do corpus.
nltk.download('stopwords')

from nltk.corpus import stopwords

stop = stopwords.words('portuguese')

for frase in nltk_dataset:
    tokens = word_tokenize(frase, language='portuguese')
    tokens_limpos = []

    for t in tokens:
        if t.lower() not in stop:
            tokens_limpos.append(t)

    print(tokens_limpos)
```

```
# Stemming simples com NLTK.
from nltk.stem import RSLPStemmer
nltk.download('rslp')
stemmer = RSLPStemmer()

for i, frase in enumerate(nltk_dataset, 1):
    tokens = word_tokenize(frase, language='portuguese')
    radicais = []

    for t in tokens:
        radicais.append(stemmer.stem(t))

    print(i, " ".join(radicais))
```

```
# Extrair e armazenar POS.
pos_data = []

for frase in spacy_dataset:
    doc = nlp(frase)
    pos_tags = [(t.text, t.pos_) for t in doc if t.is_alpha]
    pos_data.append(pos_tags)

pos_data
```

```
# Extrair e armazenar entidades nomeadas.
ner_data = []

for frase in spacy_dataset:
    doc = nlp(frase)
    entidades = [(ent.text, ent.label_) for ent in doc.ents]
    ner_data.append(entidades)

ner_data
```

### Pergunta
5. Qual abordagem tende a ser mais adequada: stemming ou lematização? Por quê?


```
# Criar versão lematizada para vetorização.
corpus_lemas = []

for frase in spacy_dataset:
    doc = nlp(frase)
    lemas = [token.lemma_ for token in doc if token.is_alpha]
    corpus_lemas.append(" ".join(lemas))

corpus_lemas
```

```
# Remover stopwords do corpus já lematizado.
corpus_sem_stop = []

for frase in spacy_dataset:
    doc = nlp(frase)
    tokens_limpos = [
        token.lemma_
        for token in doc
        if token.is_alpha and not token.is_stop
    ]
    corpus_sem_stop.append(" ".join(tokens_limpos))

corpus_sem_stop
```

## Bag of Words (BoW)

Bag of Words é uma representação vetorial esparsa em que cada documento vira um vetor com uma posição por termo do vocabulário. O valor em cada posição é tipicamente a contagem de ocorrências do termo no documento. A ideia é transformar texto em números para alimentar algoritmos de ML clássicos (classificação, clustering, similaridade), ignorando ordem e contexto.

Na prática, no scikit-learn isso é implementado pelo CountVectorizer, que converte uma coleção de documentos em uma matriz de contagem de tokens (geralmente esparsa). Você pode acessar a documentação em:

https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```
# Calcular vocabulário original sem normalização.
import re

tokens_raw = []
for frase in dataset:
    tokens_raw.extend(re.findall(r'\b\w+\b', frase))

vocab_raw = set(tokens_raw)
len(vocab_raw)
```

```
# Calcular vocabulário original com normalização básica (lowercase).
tokens_lower = [t.lower() for t in tokens_raw]
len(set(tokens_lower))
```

### Pergunta
6. Quantas linhas e quantas colunas possuem a matriz gerada? O que justifica o formato atual após a vetorização?
7. Qual é a dimensão do vocabulário gerado?

```
# Gerar BoW (matriz documento-termo) a partir do corpus lematizado.
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus_sem_stop)

feature_names = vectorizer.get_feature_names_out()

X.shape, feature_names[:]
```

```
# Visualizar a matriz BoW em formato tabular (denso) para inspeção.
bow_df = pd.DataFrame(X.toarray(), columns=feature_names)
bow_df
```

```
# Inspecionar a coluna "banco" e quais documentos a ativam.
for i in bow_df[bow_df["banco"] > 0].index:
    print(dataset[i])
```

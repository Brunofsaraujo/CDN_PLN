## Questão 01 (1.0)
a) Explique o que é uma representação esparsa em PLN e por que a codificação one-hot é considerada esparsa.

b) Na célula abaixo, implemente em Python a codificação one-hot para a lista ["casa", "carro", "filme", "bom", "ruim"] utilizando OneHotEncoder do sklearn e exiba a matriz resultante.

### RESPOSTA

Item (1A) 
Uma representação esparsa é aquela em que a matriz ou o vetor correspondente possui a esmagadora maioria de seus elementos iguais a zero (0) e apenas alguns poucos elementos diferentes de zero. A codificação One-Hot é considerada extremamente esparsa porque, se tivermos um vocabulário de tamanho $N$ (por exemplo, 10.000 palavras), cada palavra será representada por um vetor de tamanho $N$ contendo um único "1" na posição correspondente ao índice da palavra, e todo o restante (9.999 posições) preenchido com "0".

```python
# RESPOSTA Item (1B)
# Seu código abaixo:

import numpy as np
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Lista de palavras transformada em um array 2D, formato exigido pelo sklearn
palavras = np.array(["casa", "carro", "filme", "bom", "ruim"]).reshape(-1, 1)

# Inicializando o encoder (sparse_output=False retorna um array NumPy denso para visualização)
encoder = OneHotEncoder(sparse_output=False)
matriz_one_hot = encoder.fit_transform(palavras)

# Exibindo em um DataFrame para facilitar a visualização
df_one_hot = pd.DataFrame(matriz_one_hot, columns=encoder.categories_[0], index=palavras.flatten())
print("Matriz One-Hot Resultante:\n")
print(df_one_hot)
```

## Questão 02 (1.0)

a) Explique o que é stemming e qual é seu objetivo no processamento de linguagem natural.

b) Utilizando o NLTK e o RSLPStemmer, aplique stemming ao texto abaixo:
"Os pesquisadores analisavam relatórios e comparavam resultados financeiros."

### RESPOSTA

Item (2A)
Stemming é o processo heurístico de redução das palavras ao seu radical (stem), cortando sufixos e prefixos morfológicos (ex: plurais, conjugações de tempo verbal, diminutivos) sem levar em conta o contexto gramatical profundo da frase. O seu principal objetivo no PLN é reduzir a dimensionalidade do vocabulário, garantindo que variações de uma mesma palavra (como "pesquisador", "pesquisadores", "pesquisa") sejam contadas e agrupadas sob o mesmo identificador computacional comum, simplificando os dados.

```python
# RESPOSTA Item (2B)
# Seu código abaixo:

import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import RSLPStemmer

# Descomente a linha abaixo caso não tenha feito o download no ambiente:
# nltk.download('rslp')
# nltk.download('punkt')

texto = "Os pesquisadores analisavam relatórios e comparavam resultados financeiros."
tokens = word_tokenize(texto, language='portuguese')

stemmer = RSLPStemmer()
tokens_com_stemming = [stemmer.stem(t) for t in tokens]

print("Tokens Originais:", tokens)
print("Tokens (Stemming):", tokens_com_stemming)
```

## Questão 03 (1.0)

a) Explique o que são stopwords em PLN e qual é o objetivo de removê-las durante o processamento de texto.

b) Utilizando o NLTK, realize a remoção de stopwords do texto abaixo e exiba os tokens restantes:
"Os pesquisadores analisavam os dados e comparavam os resultados com os modelos anteriores."

### RESPOSTA

Item (3A)
Stopwords são palavras estruturais e de alta frequência em um idioma (como preposições, artigos e pronomes: "de", "a", "o", "que") que geralmente possuem pouco valor semântico próprio. O objetivo de removê-las é eliminar o "ruído" do texto, reduzindo o tamanho dos vetores numéricos e permitindo que os algoritmos de máquina foquem apenas nas palavras que carregam o conteúdo real (substantivos, verbos, adjetivos).

```python
# RESPOSTA Item (3B)
# Seu código abaixo:

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Descomente a linha abaixo caso não tenha feito o download
# nltk.download('stopwords')

texto = "Os pesquisadores analisavam os dados e comparavam os resultados com os modelos anteriores."
tokens = word_tokenize(texto, language='portuguese')

# Carregando as stopwords em português
stop_pt = set(stopwords.words('portuguese'))

# Filtrando
tokens_restantes = [t for t in tokens if t.lower() not in stop_pt]

print("Tokens restantes após remoção de stopwords:")
print(tokens_restantes)
```

### RESPOSTA

Item (4A)
A lematização é um processo de normalização que utiliza regras linguísticas, dicionários (tabelas de busca) e o contexto morfológico/sintático (POS tagging) para reduzir a palavra à sua forma original de dicionário, chamada "lema" (ex: "estavam" $\rightarrow$ "estar"). A principal diferença é que o stemming corta pedaços da palavra cegamente gerando muitas vezes termos que não existem no idioma (ex: "estatísticos" $\rightarrow$ "estatís"), enquanto a lematização entende o contexto e devolve uma palavra linguisticamente válida.

## Questão 05 (1.0)

a) Explique o que é o modelo Bag of Words (BoW) e qual sua principal limitação em tarefas de PLN.

b) Utilizando o CountVectorizer do sklearn, aplique BoW ao conjunto de textos abaixo e exiba a matriz resultante e o vocabulário:
textos = [
    "o sistema falhou",
    "o sistema foi atualizado",
    "falha no sistema"
]

### RESPOSTA

Item (5A)
Bag of Words (Saco de Palavras) é um modelo de representação numérica que converte texto na contagem de frequência das palavras do vocabulário que aparecem em cada documento. A principal limitação do BoW é a total perda do sentido sintático, da ordem e do contexto temporal. O modelo ignora a estrutura da frase ("O cachorro mordeu o homem" e "O homem mordeu o cachorro" geram o exato mesmo vetor) e sofre de altíssima dimensionalidade e esparsidade.

```python
# RESPOSTA Item (5B)
# Seu código abaixo:

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

textos = [
    "o sistema falhou",
    "o sistema foi atualizado",
    "falha no sistema"
]

vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(textos)
vocabulario = vectorizer.get_feature_names_out()

df_bow = pd.DataFrame(
    X_bow.toarray(), 
    columns=vocabulario, 
    index=[f"Texto {i+1}" for i in range(len(textos))]
)

print("Vocabulário (Tokens):", vocabulario)
print("\nMatriz Bag of Words:")
display(df_bow) if 'display' in globals() else print(df_bow)
```

## Questão 06 (1.5)

a) Explique o que é TF-IDF e qual é sua principal ideia em relação à importância das palavras em um conjunto de documentos.

b) Compare TF-IDF com Bag of Words (BoW), destacando a principal diferença entre as duas abordagens.

c) Utilizando o TfidfVectorizer do sklearn, aplique TF-IDF ao conjunto de textos abaixo e exiba a matriz resultante e o vocabulário:

textos = [
    "o sistema falhou",
    "o sistema foi atualizado",
    "falha no sistema"
]

### RESPOSTA

Item (6A)
TF-IDF (Frequência do Termo - Inverso da Frequência nos Documentos) é um método de cálculo de peso. A sua ideia principal é que, para analisar a importância de uma palavra para um documento específico, não basta apenas contar quantas vezes ela aparece (TF). É preciso penalizá-la caso ela seja comum em todos os outros documentos do corpus (IDF). Isso destaca as palavras discriminativas/raras.

### RESPOSTA

Item (6B)
Ambos ignoram a ordem das palavras (são modelos estatísticos). A principal diferença é que o BoW gera valores inteiros baseados na contagem bruta, o que dá vantagem excessiva para palavras comuns, gerando ruído. O TF-IDF resolve isso gerando pesos em pontos flutuantes; ele reescala a matriz BoW dando menos peso para palavras genéricas presentes em toda a coleção, evidenciando o real assunto do documento.

```python
# RESPOSTA Item (6C)
# Seu código abaixo:

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

textos = [
    "o sistema falhou",
    "o sistema foi atualizado",
    "falha no sistema"
]

tfidf_vec = TfidfVectorizer()
X_tfidf = tfidf_vec.fit_transform(textos)
vocabulario_tfidf = tfidf_vec.get_feature_names_out()

df_tfidf = pd.DataFrame(
    X_tfidf.toarray(), 
    columns=vocabulario_tfidf, 
    index=[f"Texto {i+1}" for i in range(len(textos))]
)

print("Vocabulário:", vocabulario_tfidf)
print("\nMatriz TF-IDF:")
display(df_tfidf) if 'display' in globals() else print(df_tfidf)
```

### RESPOSTA

Item (7A)
O PMI é uma métrica teórica da probabilidade que mede a "Informação Mútua" entre duas palavras. Ele avalia o quão frequentemente duas palavras ocorrem juntas comparado à probabilidade de ambas ocorrerem se fossem eventos totalmente ao acaso. A finalidade no PLN é descobrir associações semânticas e collocations fortes (como "Inteligência Artificial" ou "bebe" $\rightarrow$ "água") filtrando a relação acima da pura distribuição aleatória.

### RESPOSTA

Item (7B)
O PMI é uma métrica teórica da probabilidade que mede a "Informação Mútua" entre duas palavras. Ele avalia o quão frequentemente duas palavras ocorrem juntas comparado à probabilidade de ambas ocorrerem se fossem eventos totalmente ao acaso. A finalidade no PLN é descobrir associações semânticas e collocations fortes (como "Inteligência Artificial" ou "bebe" $\rightarrow$ "água") filtrando a relação acima da pura distribuição aleatória.

```python
# RESPOSTA Item (7C)
# Seu código abaixo:

import numpy as np

def calcular_pmi(p_wc, p_w, p_c):
    # Proteção matemática para evitar divisão por zero ou log(0) que resulta em infinitos
    if p_wc == 0 or p_w == 0 or p_c == 0:
        return 0.0
        
    return np.log2(p_wc / (p_w * p_c))

# Testando com os valores do exercício
p_gato_leite = 0.1
p_gato = 0.2
p_leite = 0.5

resultado_pmi = calcular_pmi(p_gato_leite, p_gato, p_leite)
print(f"O valor de PMI para os dados fornecidos é: {resultado_pmi}")
```

## Questão 08 (1.0)

a) Explique o que representa uma matriz palavra-contexto baseada em PMI e qual seu objetivo em PLN.

b) Utilizando Python, construa uma matriz palavra-contexto com PMI para o conjunto de textos abaixo. Considere como contexto as palavras que aparecem no mesmo documento.

textos = [
    "gato bebe leite",
    "cachorro bebe água",
    "gato come peixe"
]

### RESPOSTA

Item (8A)
A matriz de coocorrência (palavra-contexto) baseada no PMI substitui a contagem bruta entre duas palavras vizinhas na mesma frase pela sua Informação Mútua. O grande objetivo em PLN é a transição para a Semântica Distribucional: em vez de avaliar as palavras com vetores gigantes baseados em documentos esparsos, passamos a avaliá-las com vetores baseados nos vizinhos com os quais se associam ativamente. Palavras de contexto semelhante assumirão vetores mais parecidos no final.

```python
# RESPOSTA Item (8B)
# Seu código abaixo:

import pandas as pd
import numpy as np

textos = [
    "gato bebe leite",
    "cachorro bebe água",
    "gato come peixe"
]

# 1. Separar tokens e extrair o vocabulário unificado e ordenado
tokens_doc = [texto.split() for texto in textos]
vocabulario = sorted(set(word for tokens in tokens_doc for word in tokens))

# 2. Criar a matriz de contagem de coocorrências (Assumindo como contexto = MESMO documento)
matriz_contagem = pd.DataFrame(0.0, index=vocabulario, columns=vocabulario)

# Para cada documento, incrementamos cruzamentos de todas palavras com todas
for tokens in tokens_doc:
    for w in tokens:
        for c in tokens:
            matriz_contagem.loc[w, c] += 1

# 3. Probabilidades 
# O "N" é a soma de todos os eventos registrados dentro do contexto da matriz
N = matriz_contagem.values.sum()
p_wc = matriz_contagem / N                  # Probabilidade Conjunta
p_w = matriz_contagem.sum(axis=1) / N       # Probabilidade Marginal (linha)
p_c = matriz_contagem.sum(axis=0) / N       # Probabilidade Marginal (coluna)

# 4. Calcular o PMI baseando-se na fórmula (com proteção aos NaNs/Infs que dão quando Prob = 0)
import warnings
warnings.filterwarnings('ignore') # Omitindo aviso de divisão por zero local da biblioteca Pandas

pmi_matrix = np.log2(p_wc / (p_w.values[:, None] * p_c.values[None, :]))

# Converter os infinites e NaNs (gerados de log(0)) para 0 (que simula o Positive PMI / PPMI)
pmi_matrix = pmi_matrix.replace([np.inf, -np.inf, np.nan], 0.0)

print("Matriz Palavra-Contexto de PMI:")
display(pmi_matrix) if 'display' in globals() else print(pmi_matrix)
```
